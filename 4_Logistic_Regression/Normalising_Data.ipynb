```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv("Dataset/linearX.csv")
dfy = pd.read_csv("Dataset/linearY.csv")
x = dfx.values
y = dfy.values

plt.scatter(x, y)
plt.show()
```

```python
x2 = ((x - x.mean())/x.std())

plt.scatter(x, y, label = 'Raw Data')
plt.scatter(x2, y, color='orange', label = "Normalised Data")
plt.legend()
plt.show()
```

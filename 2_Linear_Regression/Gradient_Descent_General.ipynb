```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
print(X)
```

```python
Y = (X-5)**2
print(Y)
```

### Given a Function f(x), we want to find the values of x that minimizes f

```python
## Visualisation
```

```python
plt.style.use('seaborn')
plt.plot(X, Y)
plt.ylabel("F(x)")
plt.xlabel("x")
plt.show()
```

```python
x = 0
lr = 0.1
error = []
# Take 50 steps in downhill direction
plt.plot(X, Y)
for i in range(50):
    grad = 2 * (x-5)
    x = x - (lr * grad)
    y = (x-5) ** 2
    error.append(y)
    plt.scatter(x,y)
```

```python
# Plot the values of error

plt.plot(error)
plt.show()
```

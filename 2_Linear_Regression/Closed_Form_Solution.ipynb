# Closed Form Solution in Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Generate Dataset

X,Y = make_regression(n_samples=400, n_features=1, n_informative=1, noise=1.8, random_state=11)

print(X.shape, Y.shape)
```

```python
Y = Y.reshape((-1,1))
print(X.shape, Y.shape)
```

```python
# Normalise
X = (X - X.mean())/X.std()
```

```python
# Visualise
plt.figure()
plt.scatter(X, Y)
plt.title("Normalized Data")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))
X_ = np.hstack((X, ones))
print(X_.shape)
```

```python
def predict(X, theta):
    return np.dot(X, theta)
```

```python
def getThetaClosedForm(X, Y):
    Y = np.mat(Y)
    firstPart = np.dot(X.T, X)
    secondPart = np.dot(X.T, Y)
    theta = np.linalg.pinv(firstPart) * secondPart
    return theta
```

```python
theta = getThetaClosedForm(X_, Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X, Y)
plt.plot(X, predict(X_, theta), color='red', label='prediction')
plt.title("Closed Form Solution - Linear Regression")
plt.legend()
plt.show()
```
